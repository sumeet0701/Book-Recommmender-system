# Popularity based Recommender System 

```python
# only those book which having min of 250 rating give on that book
# displaying top 50 book whose average rating 
ratings = rating.merge(book,on = 'ISBN')
```

```python
ratings.head()
```

```python
num_rating = ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating.rename(columns={'Book-Rating':'Num_rating'},inplace =True)
```

```python
num_rating
```

```python
avg_rating = ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace =True)
avg_rating
```

```python
popular = avg_rating.merge(num_rating,on ='Book-Title')
```

```python
popular = popular[popular['Num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular = popular.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-S','Num_rating','avg_rating']]
```

```python
popular
```

# Collaborative Filtering Based System

```python
# step1 = we will find the user_id which has given miniumn 200 votes 
# the person whose has given atleast 200 votes in total that we will consider 
# step 2 = we will find the book_title which has given miniumn 50 votes 
```

```python
x =ratings.groupby('User-ID').count()['Book-Rating']>=200
read = x[x].index
```

```python
filtered_rating = ratings[ratings['User-ID'].isin(read)]
```

```python
filtered_rating.head()
```

```python
# On the bases of books :
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
final_ratings.duplicated().sum()
```

```python
pt =final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(pt)
```

```python
similar.shape
```

```python
def reco(name):
    # index fetch of book name :
    index = np.where(pt.index == name)[0][0]
    distance = sorted(list(enumerate(similar[index])),key = lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in distance:
        item = []
        temp_df = book[book['Book-Title']==(pt.index[i[0]])]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
    
```

```python
reco('1984')
```
